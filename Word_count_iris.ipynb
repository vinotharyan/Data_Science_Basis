```
text_file = sc.textFile("/FileStore/tables/iris.txt")
counts = text_file.flatMap(lambda line: line.split(","))\
         .map(lambda word:(word,1))\
         .reduceByKey(lambda a,b:a+b)
counts.count()
```

```
# 1.Remove words count with occurance less than 3.
cd3=counts.sortBy(lambda (word, count): count>=4).filter(lambda (word, count): count>=4)
print cd3.count()
cd3.collect()
#cd3.count()
```

```
# 1.Remove words  count with occurance less than 2.
cd4=counts.sortBy(lambda (word, count): count>=3).filter(lambda (word, count): count>=3)
print cd4.count()
cd4.collect()
#cd3.count()
```

```
# Sort the words based  in reverse order
sort = counts.sortByKey(False)
sort.collect()

```
